```python
# import sys
# !{sys.executable} -m pip install bs4
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
%matplotlib inline
```

```python
url = f"https://www.dsebd.org/dseX_share.php"
response = requests.get(url)
html = response.content
soup = bs(html)
dsex_table = soup.find_all('table', {"class":"table table-bordered background-white shares-table"})[0]
dsex_table_row = dsex_table.find_all('a')
dsex_table_row

tickers = [ ticker.text.strip() for ticker in dsex_table_row]
tickers
```

```python
df = pd.DataFrame(columns=['Ticker','Market Cap(mn)', 'Year', 'EPS', 'NAV', 'ROE'])
df['Ticker'] = tickers
df
```

```python
# Function to scrape data for a symbol
def scrape_data(symbol):
    url = f"https://www.dsebd.org/displayCompany.php?name={symbol}"
    response = requests.get(url)
    html = response.content
    soup = bs(html)
    mktcap_table = soup.find_all('table', id="company")[1]
    mktcap = float(mktcap_table.find_all('td')[-1].text.strip().replace(',','')) #replacing comma to convert str to float
    
    table = soup.find_all('table', id="company")[7]
    table_row = table.find_all('tr')[-1]
    year = int(table_row.find_all('td')[0].text.strip())
    
    try:
        eps = float(table_row.find_all('td')[4].text.strip())
        nav = float(table_row.find_all('td')[7].text.strip())
        if not eps<0 and not nav<0:
            roe = round(eps / nav, 4)  # keeping 4 decimal places
        else: roe = 0
    except ValueError:
        # Handle the ValueError (could not convert string to float)
        eps, nav, roe = None, None, None
    
    return mktcap, year, eps, nav, roe
```

```python
# Loop through the DataFrame
for i in range(len(df)):
    symbol = df['Ticker'][i]
    
    # Scraping data and update DataFrame
    mktcap, year, eps, nav, roe = scrape_data(symbol)
    print(f'i:{i}, mktcap: {mktcap}, year: {year}, eps: {eps}, nav: {nav}, roe: {roe}')
    df.at[i, 'Market Cap(mn)'] = mktcap
    df.at[i, 'Year'] = year
    df.at[i, 'EPS'] = eps
    df.at[i, 'NAV'] = nav
    df.at[i, 'ROE'] = roe
    
    # Introducing a delay after every 2 requests
    if i % 2 == 1 and i < len(df) - 1:
        time.sleep(5)  # Sleep for 5 seconds to ensure maximum 2 requests are made per 5 seconds
```

```python
df = df[df['ROE']>0].reset_index(drop=True).copy()
df_sorted = df.sort_values(by='ROE', ascending=False).reset_index(drop=True)
df_sorted
```

```python
df_sorted.to_csv('datas/DSEX_Growth.csv', index=False)
```
