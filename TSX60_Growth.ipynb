```python
def get_net_income_TTM(income_stmt):
    net_income_Q1 = income_stmt.iloc[income_stmt.index.get_loc('NetIncome'), 0]
    net_income_Q2 = income_stmt.iloc[income_stmt.index.get_loc('NetIncome'), 1]
    net_income_Q3 = income_stmt.iloc[income_stmt.index.get_loc('NetIncome'), 2]
    net_income_Q4 = income_stmt.iloc[income_stmt.index.get_loc('NetIncome'), 3]
    # print(net_income_Q1, net_income_Q2, net_income_Q3, net_income_Q4)
    net_income_TTM = net_income_Q1 + net_income_Q2 + net_income_Q3 + net_income_Q4 
    return net_income_TTM
```

```python
def get_StockholdersEquity_TTM(balance_sheet):
    StockholdersEquity_Q1 = balance_sheet.iloc[balance_sheet.index.get_loc('StockholdersEquity'), 0]
    StockholdersEquity_Q2 = balance_sheet.iloc[balance_sheet.index.get_loc('StockholdersEquity'), 1]
    StockholdersEquity_Q3 = balance_sheet.iloc[balance_sheet.index.get_loc('StockholdersEquity'), 2]
    StockholdersEquity_Q4 = balance_sheet.iloc[balance_sheet.index.get_loc('StockholdersEquity'), 3]
    # print(StockholdersEquity_Q1,StockholdersEquity_Q2,StockholdersEquity_Q3,StockholdersEquity_Q4)
    StockholdersEquity_TTM = (StockholdersEquity_Q1 + StockholdersEquity_Q2 + StockholdersEquity_Q3 + StockholdersEquity_Q4)/4
    return StockholdersEquity_TTM
```

```python
def get_TotalAssets_TTM(balance_sheet):
    TotalAssets_Q1 = balance_sheet.iloc[balance_sheet.index.get_loc('TotalAssets'), 0]
    TotalAssets_Q2 = balance_sheet.iloc[balance_sheet.index.get_loc('TotalAssets'), 1]
    TotalAssets_Q3 = balance_sheet.iloc[balance_sheet.index.get_loc('TotalAssets'), 2]
    TotalAssets_Q4 = balance_sheet.iloc[balance_sheet.index.get_loc('TotalAssets'), 3]
    # print(TotalAssets_Q1,TotalAssets_Q2, TotalAssets_Q3,TotalAssets_Q4)
    TotalAssets_TTM = (TotalAssets_Q1 + TotalAssets_Q2 + TotalAssets_Q3 + TotalAssets_Q4)/4
    return TotalAssets_TTM
```

```python
from requests import Session
from requests_cache import CacheMixin, SQLiteCache
from requests_ratelimiter import LimiterMixin, MemoryQueueBucket
from pyrate_limiter import Duration, RequestRate, Limiter
class CachedLimiterSession(CacheMixin, LimiterMixin, Session):
    pass

session = CachedLimiterSession(
    limiter=Limiter(RequestRate(2, Duration.SECOND*5)),  # max 2 requests per 5 seconds
    bucket_class=MemoryQueueBucket,
    backend=SQLiteCache("yfinance.cache"),
)
```

```python
for i in range(60):
    symbol = df['Symbol'][i]
    ticker = yf.Ticker(symbol, session=session)
    income_stmt = ticker.get_income_stmt(freq='quarterly')
    balance_sheet = ticker.get_balance_sheet(freq='quarterly')
    
    net_income_TTM = get_net_income_TTM(income_stmt)
    StockholdersEquity_TTM = get_StockholdersEquity_TTM(balance_sheet)
    print(StockholdersEquity_TTM)
    TotalAssets_TTM = get_TotalAssets_TTM(balance_sheet)
    
    df['ROA_TTM'][i]= net_income_TTM/TotalAssets_TTM
    if StockholdersEquity_TTM == 0:
        df['ROE_TTM'][i] = [None]
    else:
        df['ROE_TTM'][i]= net_income_TTM/StockholdersEquity_TTM
```
